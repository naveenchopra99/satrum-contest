```python
%%time
path = 'https://snob.ru/entry/'
start = 173000
end = 173010
# multiple_sync_requests.py
import requests

def main():  
    url = path
    session = requests.Session()
    for i in range(start, end+1):
        urlfull=url+str(i)
        print(f"making request {i} to {urlfull}")
        resp = session.get(urlfull)
        if resp.status_code != 200:
            print(resp.status_code)
        else:
            with open('data/{}.html'.format(i), 'w') as file:
                file.write(resp.text)

main() 
```

```python
%%time
#-https://stackabuse.com/asynchronous-vs-synchronous-python-performance-analysis/
#https://pawelmhm.github.io/asyncio/python/aiohttp/2016/04/22/asyncio-aiohttp.html
path = 'https://snob.ru/entry/'
start = 173000
end = 173999

#!/usr/local/bin/python3.5
import asyncio
from aiohttp import ClientSession
import aiofiles
import json

pages = []
try:
    with open('pages.txt', 'r') as f:
        pages=json.load(f)
except Exception as error:
    print(error)
print(len(pages))

async def fetch(url, i, session):
    async with session.get(url) as response:
        #data = await response.read()
        #print('read',i)
        data = await response.text(encoding='utf-8', errors='replace')
    async with aiofiles.open('data/{}.html'.format(i), 'w', encoding='utf-8', errors='replace') as file:
        #print('write',i)
        await file.write(data)
        return [url, data]

async def run():
    url = path+'{}'
    tasks = []
    d_count=0
    # Fetch all responses within one Client session,
    # keep connection alive for all requests.
    async with ClientSession() as session:
        for i in range(start, end+1):
            if str(i) not in pages:
                task = asyncio.ensure_future(fetch(url.format(i), i, session))
                tasks.append(task)
            else:
                #print('page {} already downloaded'.format(i))
                d_count+=1
        #await asyncio.gather(*tasks)
        responses = await asyncio.gather(*tasks)
        # you now have all response bodies in this variable
    
    print('download:',len(responses))
    print(' not download:',d_count)
    #print(type(responses))
    for response in responses:
        j = response[0].split('/')[-1]
        print(response[0], len(response[1]))
        pages.append(j)
        #with open('data/{}.html'.format(j), 'w') as file:
        #    file.write(response[1])

        
'''
def print_responses(result):
    print(len(result))

def save_responses(result):
    with open('data/{}.html'.format(result[0].split('/')[-1]), 'a') as file:
        file.write(str(result[1]))
'''

loop = asyncio.get_event_loop()
future = asyncio.ensure_future(run())
loop.run_until_complete(future)

try:
    with open('pages.txt', 'w') as f:
        json.dump(pages,f)
except Exception as error:
    print(error)
print(len(pages))
```
